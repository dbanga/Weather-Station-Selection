```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# importing temp data 

temp_data=pd.read_csv('temp_hist.csv')

temp_data['Date'] = pd.to_datetime(temp_data['Date'])
temp_data['year'], temp_data['month']= temp_data['Date'].dt.year, temp_data['Date'].dt.month


temp_data['week'] = pd.to_datetime(temp_data['Date'], format='%m%d%Y').dt.strftime('%W').astype(int)

temp_data=temp_data.drop('Date', axis=1)

temp_data['T2']=temp_data['temp']**2
temp_data['T3']=temp_data['temp']**3
temp_data['Hour*temp']=temp_data['temp']*temp_data['Hour']
temp_data['Hour*T2']=temp_data['temp']*temp_data['Hour']
temp_data['Hour*T3']=temp_data['temp']*temp_data['Hour']
temp_data['Week*temp']=temp_data['temp']*temp_data['week']
temp_data['Week*T2']=temp_data['temp']*temp_data['week']
temp_data['Week*T3']=temp_data['temp']*temp_data['week']
temp_data['Week*Hour']=temp_data['Hour']*temp_data['week']

# splitting the data into  training and testing
temp_data_train=temp_data[(temp_data['year'].isin([2005,2006]))] # training data
temp_data_test=temp_data[(temp_data['year']==2007)] # testing data
```

```python
# importing temp data and Load data 
Load_data=pd.read_csv('load_hist.csv')

Load_data['Date'] = pd.to_datetime(Load_data['Date'])
Load_data['year'], Load_data['month'] = Load_data['Date'].dt.year, Load_data['Date'].dt.month

Load_data['week'] = pd.to_datetime(Load_data['Date'], format='%m%d%Y').dt.strftime('%W').astype(int)

Load_data=Load_data.drop('Date', axis=1)

# splitting the data into  training and testing
Load_data_train=Load_data[(Load_data['year'].isin([2005,2006]))] #training data
Load_data_test=Load_data[(Load_data['year']==2007)] # testing data

Load_data.describe()
```

```python
# using WS as filter, generating the independent data for various WS by changing it manually 

temp_data_train_X= temp_data_train[(temp_data_train['WS']==12)] #manually change the number of WS(1-28)
temp_data_train_X=temp_data_train_X.drop('WS', axis=1)
temp_data_train_X.describe()
```

```python
# runninng the linear regression and predit the training period

import statsmodels.api as sm

model = sm.OLS(np.asarray(Load_data_train['Load']), np.asarray(temp_data_train_X)).fit()

Load_train_predict= model.predict(temp_data_train_X)
model.summary()
```

```python
#note the mape of traning period for each WS ( in excel)

from sklearn.metrics import mean_squared_error
from math import sqrt
def mean_absolute_percentage_error(Y_test, Y_pred): 
    Y_test, Y_pred = np.array(Y_test), np.array(Y_pred)
    return np.mean(np.abs((Y_test - Y_pred) / Y_test)) * 100

mae_train=mean_absolute_percentage_error(Load_data_train['Load'], Load_train_predict)
print(mae_train)

```
